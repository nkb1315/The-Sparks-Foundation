We shall create a function to plot the count plot for Categorical Variables.

```python
#creating a function for plotting the countplot for Categorical Variables

def countplot(i):
    plt.figure(figsize = [10,5])
    plt.style.use('ggplot')

    ax = sns.countplot(data = df, x = i, palette = "Set1")
    total=len(df[i])
    
    plt.xticks()


    #Adding the title
    plt.title("Countplot of {}".format(i), fontdict={'fontsize': 16, 'fontweight' : 12, 'color' : 'Maroon'})

    for p in ax.patches:
        text = '{:.1f}%'.format(100*p.get_height()/total)
        x = p.get_x() + p.get_width() / 2.
        y = p.get_height()
        
        ax.annotate(text, (x,y), ha = 'center', va = 'center', xytext = (0, 5), textcoords = 'offset points')
    

    plt.show()
```

### We shall combine columns and see the overall Sales and Profit across different features.


#### Sales Analysis across Regions:

```python
region_sales = df.groupby(by ='Region')['Sales'].sum().sort_values(ascending = False)
region_sales
```

```python
ax = region_sales.plot(kind = 'bar',figsize=(15,8),title='Total Sales across Region',color = 'gold')

for i in ax.containers:
    ax.bar_label(i,)

plt.ylabel("Sales in USD")

plt.show()
```

**Observations**:
- West region has the highest Sales.
- South Region has the lowest sales.

#### Profit Analysis across Regions:

```python
region_profit = df.groupby(by ='Region')['Profit'].sum().sort_values(ascending = False)
region_profit
```

```python
ax = region_profit.plot(kind = 'bar',figsize=(15,8),title='Total Profit across Region',color = 'yellowgreen')

for i in ax.containers:
    ax.bar_label(i,)

plt.ylabel("Profit in USD")

plt.show()
```
